Um notebook dedicado para a simulação de Processos Poisson.

```python
import scipy.stats as stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Fregueses chegam a uma taxa de $\lambda=2$ por minuto

```python
exponential_time = stats.gamma.rvs(a=1, scale=1/2,size=10)
gamma_time = np.cumsum(exponential_time)

fig, ax = plt.subplots(figsize=(10,6))

N = np.arange(1, len(gamma_time)+1)

poi = 'Poisson'

ax.step(gamma_time, N, label=f'$N(t) \sim Poisson(\lambda=2)$')
ax.set_ylabel(f'$N(t)$')
for i,t in enumerate(gamma_time):
    index = i+1
    ax.text(x=t, y = 0.5, s=f'$W_{index}$', horizontalalignment='center',
     verticalalignment='center')

ax.set_title(f'Realização de um processo de Poisson.')
ax.grid(True)
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(12,8))

N = np.arange(1, 11+1)

for i in range(5):
    exponential_time = stats.expon.rvs(scale=1/0.25, size=10)
    gamma_time = np.cumsum(exponential_time)
    gamma_time = np.concatenate(([0.0], gamma_time))
    ax.step(gamma_time, N)

ax.set_ylabel(f'$N(t)$')
ax.set_xlabel(f'$W_n$')
ax.set_title(f'Processos de Poisson')
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.grid(True)
```

# Processos de Poisson Generalizados

O modelo de um processo de poisson como uma combinação linear de processos de poisson e o processo de poisson composto.

```python
T = 15.0 # O tempo final.
lam = 0.5 # taxa que os eventos ocorrem.

exp_time = stats.expon.rvs(scale=1/lam, size=10)
gamma_time = np.cumsum(exp_time)
gamma_time = gamma_time[gamma_time < T]
sample_size = len(gamma_time)

Y = stats.geom.rvs(p=0.25, size=sample_size)

S = np.cumsum(Y)

fig, ax = plt.subplots(figsize=(10,8))

ax.step(gamma_time, S)
```
