# Cadeia de Markov de 2 estados

É escolhido $S=\{0,1\}$ onde $P_{01} = 0.7$ e $P_{10}=0.3$, o que nos dá a matriz de transição:
$$
P = 
\begin{pmatrix}
0.3 & 0.7 \\
0.3 & 0.7
\end{pmatrix}
$$

```python
def Gerador(X0):
    if X0 == 0:
        return np.random.choice([0,1], replace=True, p = [0.7, 0.3])
    else:
        return np.random.choice([0,1], replace=True, p = [0.7, 0.3])
```

```python
def SamplePath(n=100, X0=0):
    result = [X0]
    for i in range(n):
        X1 = Gerador(X0)
        result.append(X1)
        X0 = X1
    return result
```

```python
test = SamplePath(n=200, X0=0)
n = np.arange(0, 201, 1)

plt.scatter(x=n, y=test)
```
