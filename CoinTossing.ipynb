Vaguely, almost everyone believes that for large Â«, the number of heads is about the same as the number of tails. That is, if you toss a fair coin a large number of times, then about half the tosses result in heads.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

N = 500
n = np.arange(0, N)

ax.plot(n, stats.binom.pmf(n, N, 1/2))
ax.axhline(1/np.sqrt(np.pi*N/2), color='red', label='Stirling')
ax.legend()
```
